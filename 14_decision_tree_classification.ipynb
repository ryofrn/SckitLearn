## Konsep Dasar

### Teminologi
- Root Node -> node paling awal/atas
- Leaf Node -> node paling akhir/bawah
- Internal node -> node di antara Leaf Node dan Root Node (tengah tengah)

### Gini Impurity
- cara mengukur ketidakmurnian (impurity)
- nilai gini impurity berkisar dari 0.0 sampai dengan 1.0 (0 = murni, 1 = tidak murni)
- Rumus : $
\begin{align*} 
G &= 1 - \sum_i^n P_i^2 \\
\end{align*}
$

### Information Gain
- Information gain didapat dari nilai gini impurity dikurangi mean gini impurity

## Materi

### Dataset
Sumber :https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0)
```

### Classification dengan `DecisionTreeClassifier`

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)

model.fit(X_train, y_train)
```

### Visualisasi Model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

### Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
